```python
import json
import pandas as pd
import numpy as np
from spacy.lang.en import English
```

```python
# load dataset into a dataframe

!wget -qO "Mohammad_dataset_result(3).json" https://uoguelphca-my.sharepoint.com/:u:/g/personal/mhavaled_uoguelph_ca/ET_4hgQABD5JhtiDtFGUG4gBASpIv6WURezHphMTlPCJqg?download=1
data = json.load(open('Mohammad_dataset_result(3).json'))
df = pd.DataFrame(data)
df
```

```python
# split articles into sentences
# every entry of df['sentences'] will contain a list of strings

spacy_eng_nlp = English()
spacy_eng_nlp.add_pipe("sentencizer")

df['sentences'] = df.apply(
    lambda x: (
        [str(s) for s in spacy_eng_nlp(x.paragraphs).sents]
    ),
    axis=1
)
df
```

```python
# indicate which is the key sentence

df['is_key'] = df.apply(
    lambda x: (
        [int(x.key_sentences in s) for s in x.sentences]
    ),
    axis=1
)

df['key_idx'] = df['is_key'].apply(lambda x: np.argmax(x)+1)
df['num_of_sentences'] = df['is_key'].apply(lambda x: len(x))

df
```

```python
# show all "single-sentence" paragraphs

df_issue1 = df[df.is_key.apply(len) == 1]
df_issue1
```

```python
# show all paragraphs which cannot locate one and only one key sentence

df_issue2 = df[df.is_key.apply(sum)!=1]
df_issue2
```

```python
# show remaining paragraps

df_remain = df[~df.index.isin(df_issue1.index.tolist() + df_issue2.index.tolist())]
df_remain
```

```python
# count how many have key_sentence at the last sentence

(df_remain.key_idx == df_remain.num_of_sentences).sum()
```
